#### Randomly select 50 reviews from different businesses with rating 1

```python
#uniqueBusiness = data['business_id'].unique()
#uniqueBusiness.size
```

```python
# Filter reviews with 1 star and store in a dataframe
oneStar = data[data['stars']==1]
```

```python
# Group by business ID and sample a random 1-star review from each group (1 per business)
grouped = oneStar.groupby('business_id').sample(n=1)
len(grouped)
```

```python
# Randomly select 50 reviews of rating 1 (one from each business)
random50 = grouped.sample(n=50)
#len(random50)
display(random50)
```

#### Extract all noun-adjective pairs

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
noun_adj_pairs = []

for i in random50['text']:
  doc = nlp(i)
  for i,token in enumerate(doc):
      if token.pos_ not in ('NOUN','PROPN'):
          continue
      for j in range(i+1,len(doc)):
          if doc[j].pos_ == 'ADJ':
              noun_adj_pairs.append((token,doc[j]))
              break
noun_adj_pairs
```

#### Find the top 10 most frequent noun-adjective pairs

```python
# (i found a few diff methods online to find the most frequent pair)
# method 1
 
def most_frequent(List):
    counter = 0
    num = List[0]
     
    for i in List:
        curr_frequency = List.count(i)
        if(curr_frequency> counter):
            counter = curr_frequency
            num = i
 
    return num
 
print(most_frequent(noun_adj_pairs))
```

```python
# method 2
def most_common(lst):
    return max(set(lst), key=lst.count)
print(most_common(noun_adj_pairs))
```

```python
# method 3
import itertools
import operator

def most_common(L):
  # get an iterable of (item, iterable) pairs
  SL = sorted((x, i) for i, x in enumerate(L))
  # print 'SL:', SL
  groups = itertools.groupby(SL, key=operator.itemgetter(0))
  # auxiliary function to get "quality" for an item
  def _auxfun(g):
    item, iterable = g
    count = 0
    min_index = len(L)
    for _, where in iterable:
      count += 1
      min_index = min(min_index, where)
    # print 'item %r, count %r, minind %r' % (item, count, min_index)
    return count, -min_index
  # pick the highest-count/earliest item
  return max(groups, key=_auxfun)[0]

most_common(noun_adj_pairs)
```
